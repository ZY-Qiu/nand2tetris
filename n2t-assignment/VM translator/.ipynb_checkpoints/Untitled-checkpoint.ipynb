```python
#use a running interger j to differentiate between functions...
func_name = 0
function_name = [0]
j = 0
#以上for 循环外

if (cmd[0] == "label"):
    if (func_name) == 0:
        new.append("({})\n".format(cmd[1]))
    else:
        new.append("({}.{}${}.{})\n".format(self.data_name, function_name[func_name], cmd[1], j))
    
if (cmd[0] == "goto"):
    if (func_name == 0):
        new.append("@{}\n".format(cmd[1]))
        new.append("0;JMP\n")
    else:
        new.append("@{}.{}${}.{}\n".format(self.data_name, function_name[func_name], cmd[1], j))
        new.append("0;JMP\n")
            
if (cmd[0] == "if-goto"):
    new.append("@SP\n")
    new.append("A = M-1\n")
    new.append("D = M\n")
    if (func_name == 0):
        new.append("@{}\n".format(cmd[1]))
        new.append("D;JNE\n")
    else:
        new.append("@{}.{}${}.{}\n".format(self.data_name, function_name[func_name], cmd[1], j))
        new.append("D;JNE\n")

if (cmd[0] == "function"):
    new.append("({})\n".format(cmd[1]))
    for a in range(int(cmd[2])):
        #push 0
        new.append('@0\n')
        new.append('D = A\n')
        new.append('@SP\n')
        new.append('M = M+1\n')
        new.append('A = M-1\n')
        new.append('M = D\n')
    
if (cmd[0] == "call"):
    #加一指示参数，并保存
    #记录当前函数名
    function_name.append()
    func_name += 1
    j += 1
    #处理栈和内存段
    new.append('@{}\n'.format(j))
    new.append('D = A\n')
    new.append('@SP\n')
    new.append('M = M+1\n')
    new.append('A = M-1\n')
    new.append('M = D\n')
    #push LCL
    new.append('@LCL\n')
    new.append('D = M\n')
    new.append('@SP\n')
    new.append('M = M+1\n')
    new.append('A = M-1\n')
    new.append('M = D\n')
    #push ARG
    new.append('@ARG\n')
    new.append('D = M\n')
    new.append('@SP\n')
    new.append('M = M+1\n')
    new.append('A = M-1\n')
    new.append('M = D\n')
    #push THIS
    new.append('@THIS\n')
    new.append('D = M\n')
    new.append('@SP\n')
    new.append('M = M+1\n')
    new.append('A = M-1\n')
    new.append('M = D\n')
    #push THAT
    new.append('@THAT\n')
    new.append('D = M\n')
    new.append('@SP\n')
    new.append('M = M+1\n')
    new.append('A = M-1\n')
    new.append('M = D\n')
    #ARG = SP-5-nArgs
    new.append('@5\n')
    new.append('D = A\n')
    new.append('@{}\n'.format(int(cmd[2])))
    new.append('D = M + D\n')
    new.append('@SP\n')
    new.append('D = M - D\n')
    new.append('@ARG\n')
    new.append('M = D\n')
    #LCL = SP
    new.append('@SP\n')
    new.append('D = M\n')
    new.append('@LCL\n')
    new.append('M = D\n')
    #goto func
    new.append("@{}\n".format(cmd[1]))
    new.append("0;JMP\n")
    #(returnAddress)
    new.append("({}.{}$ret.{})\n".format(self.data_name, function_name[func_name], j))
    
if (cmd[0] == "return"):
    #去除当前保存函数名
    function_name = function_name.pop(-1)
    func_name -= 1
    #将返回值push进栈，再goto返回位置
    #恢复之前储存的状态
    new.append('@LCL\n')
    new.append('D = M\n')
    new.append('@endFrame.{}\n'.format(j))
    new.append('M = D\n')
    
    #*ARG = pop()
    new.append('@SP\n')
    new.append('M = M-1\n')
    new.append('A = M\n')
    new.append('D = M\n')
    new.append('@ARG\n')
    new.append('A = M\n')
    new.append('M = D\n')
    #SP = ARG+1
     new.append('@ARG\n')
    new.append('D = M+1\n')
    new.append('@SP\n')
    new.append('M = D\n')
    #THAT = *(endFrame - 1)
    new.append('@endFrame.{}\n'.format(j))
    new.append('D = M-1\n')
    new.append('@THAT\n')
    new.append('M = D\n')
    #THIS = *(endFrame - 2)
    new.append('@2\n')
    new.append('D = A\n')
    new.append('@endFrame.{}\n'.format(j))
    new.append('D = M-D\n')
    new.append('@THIS\n')
    new.append('M = D\n')
    #ARG = *(endFrame - 3)
    new.append('@3\n')
    new.append('D = A\n')
    new.append('@endFrame.{}\n'.format(j))
    new.append('D = M-D\n')
    new.append('@ARG\n')
    new.append('M = D\n')
    #LCL = *(endFrame - 4)
    new.append('@4\n')
    new.append('D = A\n')
    new.append('@endFrame.{}\n'.format(j))
    new.append('D = M-D\n')
    new.append('@LCL\n')
    new.append('M = D\n')
    #goto return address
    new.append("@{}.{}$ret.{}\n".format(self.data_name, function_name[func_name], j))
    new.append('0;JMP\n')
```
