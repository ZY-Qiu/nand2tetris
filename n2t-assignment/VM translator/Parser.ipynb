```python
#Parser

class Parser():
    
    def __init__(self, data, data_name):
        self.data = data
        self.data_name = data_name
        
    def parse(self):
        new = []
        with open(self.data, 'r') as f:
            lines = f.readlines()
            for i in range(0, len(lines)):
                new.append('//' + lines[i])
                #lines[i] = lines[i].rstrip('\n') //split函数已去除所有空格和换行符和制表符
                #parse the line, 在翻译完后添加换行符\n
                cmd = lines[i].split()
                if (cmd[0] == "pop"):
                    if (cmd[1] == "local"): #int(cmd[2])
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@LCL\n')
                        new.append('M = M + D\n')
                        new.append('@SP\n')
                        new.append('M = M-1\n')
                        new.append('A = M\n')
                        new.append('D = M\n')
                        new.append('@LCL\n')
                        new.append('A = M\n')
                        new.append('M = D\n')
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@LCL\n')
                        new.append('M = M - D\n')
                        
                    if (cmd[1] == "arguement"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@ARG\n')
                        new.append('M = M + D\n')
                        new.append('@SP\n')
                        new.append('M = M-1\n')
                        new.append('A = M\n')
                        new.append('D = M\n')
                        new.append('@ARG\n')
                        new.append('A = M\n')
                        new.append('M = D\n')
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@ARG\n')
                        new.append('M = M - D\n')
                        
                    if (cmd[1] == "this"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THIS\n')
                        new.append('M = M + D\n')
                        new.append('@SP\n')
                        new.append('M = M-1\n')
                        new.append('A = M\n')
                        new.append('D = M\n')
                        new.append('@THIS\n')
                        new.append('A = M\n')
                        new.append('M = D\n')
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THIS\n')
                        new.append('M = M - D\n')
                        
                    if (cmd[1] == "that"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THAT\n')
                        new.append('M = M + D\n')
                        new.append('@SP\n')
                        new.append('M = M-1\n')
                        new.append('A = M\n')
                        new.append('D = M\n')
                        new.append('@THAT\n')
                        new.append('A = M\n')
                        new.append('M = D\n')
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THAT\n')
                        new.append('M = M - D\n')
                        
                    if (cmd[1] == "constant"):
                        new.append('//Meaningless command!\n')
                        
                    if (cmd[1] == "static"):
                        new.append('@SP\n')
                        new.append('M = M-1\n')
                        new.append('A = M\n')
                        new.append('D = M\n')
                        new.append('@{}.{}\n'.format(self.data_name, int(cmd[2])))
                        new.append('M = D\n')
                        
                    if (cmd[1] == "pointer"):
                        if (int(cmd[2]) == 0):
                            new.append('@SP\n')
                            new.append('M = M-1\n')
                            new.append('A = M\n')
                            new.append('D = M\n')
                            new.append('@THIS\n')
                            new.append('M = D\n')
                        if (int(cmd[2]) == 1):
                            new.append('@SP\n')
                            new.append('M = M-1\n')
                            new.append('A = M\n')
                            new.append('D = M\n')
                            new.append('@THAT\n')
                            new.append('M = D\n')
                            
                    if (cmd[1] == "temp"):
                        if (cmd[2] > 7):
                            print("//Meaningless parameter!")
                        else:
                            new.append('@5\n')
                            new.append('D = A\n')
                            new.append('@TEMP\n')
                            new.append('M = D\n')
                            new.append('@{}\n'.format(int(cmd[2])))
                            new.append('D = A\n')
                            new.append('@TEMP\n')
                            new.append('M = M + D\n')
                            new.append('@SP\n')
                            new.append('M = M-1\n')
                            new.append('A = M\n')
                            new.append('D = M\n')
                            new.append('@TEMP\n')
                            new.append('A = M\n')
                            new.append('M = D\n')
                        
                    
                else if (cmd[0] == "push"):
                    if (cmd[1] == "local"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@LCL\n')
                        new.append('D = M + D\n')
                        new.append('A = D\n')
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "arguement"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@ARG\n')
                        new.append('D = M + D\n')
                        new.append('A = D\n')
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "this"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THIS\n')
                        new.append('D = M + D\n')
                        new.append('A = D\n')
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "that"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@THAT\n')
                        new.append('D = M + D\n')
                        new.append('A = D\n')
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "constant"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "static"):
                        new.append('@{}.{}\n'.format(self.data_name, int(cmd[2])))
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                    if (cmd[1] == "pointer"):
                        if (int(cmd[2]) == 0):
                            new.append('@THIS\n')
                            new.append('D = M\n')
                            new.append('@SP\n')
                            new.append('M = M+1\n')
                            new.append('A = M-1\n')
                            new.append('M = D\n')
                        if (int(cmd[2]) == 1):
                            new.append('@THAT\n')
                            new.append('D = M\n')
                            new.append('@SP\n')
                            new.append('M = M+1\n')
                            new.append('A = M-1\n')
                            new.append('M = D\n')
                        
                    if (cmd[1] == "temp"):
                        new.append('@{}\n'.format(int(cmd[2])))
                        new.append('D = A\n')
                        new.append('@5\n')
                        new.append('A = A + D\n')
                        new.append('D = M\n')
                        new.append('@SP\n')
                        new.append('M = M+1\n')
                        new.append('A = M-1\n')
                        new.append('M = D\n')
                        
                else if (cmd[0] == "add"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('M = D+M\n')
                    
                else if (cmd[0] == "neg"):
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = -M\n')
                    
                else if (cmd[0] == "sub"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('M = M-D\n')
                    
                else if (cmd[0] == "lt"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('D = M-D\n')
                    new.append('@lt\n')
                    new.append('D;JLT\n')
                    new.append('@nlt\n')
                    new.append('D;JGE\n')
                    new.append('(lt)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 1\n')
                    new.append('@continue{}\n'.format(i))
                    new.append('0;JMP\n')
                    new.append('(nlt)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 0\n')
                    new.append('(continue{})\n'.format(i))
                    
                else if (cmd[0] == "eq"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('D = M-D\n')
                    new.append('@eq\n')
                    new.append('D;JEQ\n')
                    new.append('@neq\n')
                    new.append('D;JNE\n')
                    new.append('(eq)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 1\n')
                    new.append('@continue{}\n'.format(i))
                    new.append('0;JMP\n')
                    new.append('(neq)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 0\n')
                    new.append('(continue{})\n'.format(i))
                    
                else if (cmd[0] == "gt"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('D = M-D\n')
                    new.append('@gt\n')
                    new.append('D;JGT\n')
                    new.append('@ngt\n')
                    new.append('D;JLE\n')
                    new.append('(gt)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 1\n')
                    new.append('@continue{}\n'.format(i))
                    new.append('0;JMP\n')
                    new.append('(ngt)\n')
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = 0\n')
                    new.append('(continue{})\n'.format(i))
                    
                else if (cmd[0] == "and"):
                    new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('M = M&D\n')
                    
                else if (cmd[0] == "or"):
                     new.append('@SP\n')
                    new.append('M = M-1\n')
                    new.append('A = M\n')
                    new.append('D = M\n')
                    new.append('A = A-1\n')
                    new.append('M = M|D\n')
                    
                else if (cmd[0] == "not"):
                    new.append('@SP\n')
                    new.append('A = M-1\n')
                    new.append('M = !M\n')
                
            new.append('(end)\n')
            new.append('@end\n')
            new.append('0;JMP\n')
```
